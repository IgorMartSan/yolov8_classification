```python
import cv2
from ultralytics import YOLO
import torch
import time

# Load the YOLOv8 model
#print(torch.cuda.is_available())  # Verifica se a GPU está disponível
#print(torch.cuda.current_device())  # ID da GPU atual
#print(torch.cuda.device(0))  # Informações sobre a GPU atual
#print(torch.cuda.get_device_name(0))



model = YOLO('./runs/classify/train2/weights/best.pt',)
model.to('cuda')
torch.cuda.set_device(0)
device: str = "mps" if torch.backends.mps.is_available() else "cpu"
print(f"O arquivo está usando a GPU ou CPU: {device}")

video_path = ".//videos//pessima.mp4"


def play_video(video_path, largura_desejada = 800, altura_desejada = 600):
    # Open the video file
    #cap = cv2.VideoCapture("C:\\Projetos\\Igor\\Projetos\\yolov8-20230621T113157Z-001\\yolov8\\yolov8_classifier_homer_bart\\video.mp4")
    cap = cv2.VideoCapture(video_path)
    while True:
       
        if cap.isOpened():
            # Read a frame from the video
            success, frame = cap.read()
            if success:
                # Run YOLOv8 inference on the frame
                # Defina a largura e a altura desejadas para a imagem redimensionada
                # Redimensione a imagem
                frame = cv2.resize( frame, (largura_desejada, altura_desejada))

                results = model(frame, device=0)

                # Visualize the results on the frame
                annotated_frame = results[0].plot(masks = True, kpt_line=True, labels=True, boxes=True)
                print(results[0].names)
                print(results[0].probs)
              
                # Display the annotated frame
                cv2.imshow("YOLOv8 Inference", annotated_frame)
                # Break the loop if 'q' is pressed
                if cv2.waitKey(1) & 0xFF == ord("q"):
                    break
            else:
                # Break the loop if the end of the video is reached
                break


    # Release the video capture object and close the display window
    print("Tentando reconectar")
    cap.release()
    cv2.destroyAllWindows()

def play_video_and_save(video_path, output_path, largura_desejada=800, altura_desejada=600):
    # Open the video file
    cap = cv2.VideoCapture(video_path)

    # Define o codec de vídeo e o objeto VideoWriter
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter(output_path, fourcc, 20.0, (largura_desejada, altura_desejada))

    while True:
        if cap.isOpened():
            # Read a frame from the video
            success, frame = cap.read()
            if success:
                # Redimensione o quadro
                frame = cv2.resize(frame, (largura_desejada, altura_desejada))

                # Execute a inferência YOLOv8 no quadro
                results = model(frame, device=0)

                # Visualize os resultados no quadro
                annotated_frame = results[0].plot(masks=True, kpt_line=True, labels=True, boxes=True)

                # Salve o quadro no arquivo de saída
                out.write(annotated_frame)

                # Exiba o quadro anotado
                cv2.imshow("YOLOv8 Inference", annotated_frame)

                # Pressione 'q' para interromper a execução
                if cv2.waitKey(1) & 0xFF == ord("q"):
                    break
            else:
                # Encerre o loop quando o final do vídeo for alcançado
                break

    # Libere o objeto de captura de vídeo e feche a janela de exibição
    cap.release()
    out.release()
    cv2.destroyAllWindows()

# Exemplo de uso:



```
